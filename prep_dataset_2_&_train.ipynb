```python
# modify the id

import os

# Define the directory path where the text files are located
directory = '/content/training/labels'

# Define the numbers you want to replace and their replacements
numbers_to_replace = {
    "2": "0",
    "3": "1",
    "4": "2",
    "6": "3",
    "8": "4"
}

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        
        # Read the contents of the file
        with open(file_path, 'r') as file:
            file_contents = file.readlines()
        
        # Modify the contents of the file
        modified_contents = []
        for line in file_contents:
            # Check if any of the numbers need to be replaced
            for number, replacement in numbers_to_replace.items():
                if line.startswith(number):
                    # Replace the number at the start of the line
                    modified_line = line.replace(number, replacement, 1)
                    line = modified_line
                    break
            
            modified_contents.append(line)
        
        # Write the modified contents back to the file
        with open(file_path, 'w') as file:
            file.writelines(modified_contents)

print('Numbers replaced successfully in all files.')
```
